```python
import pandas as pd
import matplotlib.pyplot as plt
df_fitness_original = pd.read_csv("../data/smart_fitness_&_nutrition_analytics.csv")
df_fitness = df_fitness_original
```

Limpieza a realizar
-  1º Los nombres de las columnas, cambiare los nombres a minusculas, cambiando los espacios por barras bajas y eleminando los parentesis.
-  2º Quitar las mayusculas de los valores.
-  3º Cambiar los valores de "age" a numeros enteros.
-  4º Cambiar "fat_percentage" a valores enteros.

Cambio de nombres de columnas

```python
def change_column_name(df):
	#lo primero que voy a hacer es cambiar los nombres de las columnas para dejarlos en minusculas y cambiar weight(kg) por weight_gk, lo mismo con lo otros que tengan parentesis
	#ya que creo que es mas facil y viable de utilizar de esta manera
	new_name = ""
	for i in df:
		#pones todo en minusculas
		new_name = i.lower()
		#cambiamos los espacios por barras bajas
		new_name = new_name.replace(" ", "_")
		#quitamos los parentesis
		new_name = new_name.replace("(", "")
		new_name = new_name.replace(")", "")
		df = df.rename(columns = {i: new_name})
	return df
df_fitness = change_column_name(df_fitness)
df_fitness.sample(1)
```

Cambio nombres valores

```python
def change_value_name(df):
	#ahora quiero quitar las mayusculas a todos los valores por el mismo motivo, mas facilidad de usar
	#recorremos el data frame sacando el nombre de las columnas como column
	for column in df:
		#recorremos los indices, para ello sacamos cuantas filas hay para usarlo como recorrido, en este caso del 0 al 19999, y usamos index como el indice
		for index in range(df.shape[0]):
			#hay valores que son string y otros que no, asi que usamos este if para cambiar solamente los valores string
			if type(df.at[index, column]) == str:
				#usamos el index y el column para sacar el valor, y lo guardamos en el mismo sitio pero con lower para quitar las mayusculas
				df.at[index, column] = df.at[index, column].lower()
	return df
df_fitness = change_value_name(df_fitness)
df_fitness.sample(1)
```

Cambio valores de "age"

```python
def change_column_type_to_int(df, name):
	#vamos a cambiar la edad para que sean enteros, ya que va a poder ser un estudio mas facil y no creo que sea interesante decir que una persona tiene 34 años y medio
	df[name] = df[name].astype(int)
	return df
df_fitness = change_column_type_to_int(df_fitness, "age")
df_fitness.sample(1)
```

Cambio valores de "fat_percentage"

```python

#para este caso de estudio voy a pasar fat_percentage a int, ya que si no vamos a tener una divercidad demasiado grande con tantos decimales
df_fitness = change_column_type_to_int(df_fitness, "fat_percentage")
df_fitness["fat_percentage"].sample(2)
```

Borrado de las columnas name_of_exercise, sets, reps, benefit

```python
df_fitness = df_fitness.drop(columns = ["name_of_exercise", "sets", "reps", "benefit", "target_muscle_group", "equipment_needed", "difficulty_level", "body_part", "type_of_muscle", "workout"])
for i in df_fitness:
	print(i)
```

Vamos a guardar nuestra nueva tabla limpia

```python
df_fitness.to_csv("../data/clean_smart_fitness_&_nutrition_analytics.csv")
```
